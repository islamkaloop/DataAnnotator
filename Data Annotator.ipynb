```python
from firebase_admin import db
```

```python
Tokens = {}
for i in range(len(data)):
    Token = {
        'ID':int(data["#"][i]),
        'SentID':int(data["Sentence#"][i]),
        'Token':str(data["Token"][i]),
        'Tag':str(""),
        'MixedTag':str(""),
        'Segem':str(""),
        'SegemN':str(""),
    }
    
    if(Tokens.get(str(Token["SentID"])) == None):
        Tokens[str(Token["SentID"])] = {
            str(Token["ID"]):Token
        }
    else:
        sentTokens = Tokens.get(str(Token["SentID"]))
        sentTokens[str(Token["ID"])] = Token
        Tokens[str(Token["SentID"])] = sentTokens
        
Tokens
```

```python
data = {
    "Config":{
        "CurSentence":1,
        "CurToken":1
        },
    "Tokens":Tokens
}
```

```python
import json
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```
